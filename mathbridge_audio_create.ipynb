```python
from IPython.display import Audio
from TTS.api import TTS
from datasets import load_dataset, load_from_disk
import os
import gc
import shutil
import torch
```

```python
# Get device
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
source_dir = "/kaggle/input/mathbridge-filtered"

# Destination directory
destination_dir = "/kaggle/working/mathbridge-filtered"

shutil.copytree(source_dir, destination_dir)
```

```python
os.makedirs("/kaggle/working/speech", exist_ok=True)
```

```python
ds_train = load_from_disk("/kaggle/working/mathbridge-filtered/mathbridge_filtered")
ds_train
```

```python
model_name = "tts_models/en/ljspeech/fast_pitch"
tts = TTS(model_name).to(device)
```

```python
start = 0
for idx, row in enumerate(ds_train.select(range(start, start+10000)), start):
    print(idx)
    file_name = f"speech/tts_{idx}.mp3"
    file_path = f"/kaggle/working/{file_name}"
    tts.tts_to_file(text=row["spoken_English"], file_path=file_path)
```
